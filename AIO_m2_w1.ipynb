```python
def convert2gray(vector):
    return np.max(vector)*0.5+ np.min(vector)*0.5

import matplotlib.image as mpimg
img= mpimg.imread('dog.jpeg')
gray_img_01=  np.apply_along_axis(convert2gray, axis=2, arr= img)
gray_img_01[0,0]
```

```python
import matplotlib.pyplot as plt
plt.imshow(gray_img_01, cmap="gray")
```

```python
def convert2gray_01(vector):
    return np.mean(vector)

gray_img_02= np.apply_along_axis(convert2gray_01, axis=2, arr= img)
gray_img_02[0,0]
```

```python
plt.imshow(gray_img_02, cmap="gray")
```

```python
def convert2gray_02(vector):
    return vector[0]*0.21+ vector[1]*0.72+ vector[2]*0.07

gray_img_03= np.apply_along_axis(convert2gray_02, axis=2, arr= img)
gray_img_03[0,0]
```

```python
plt.imshow(gray_img_03, cmap="gray")
```

```python
import pandas as pd
```

```python
df= pd.read_csv("advertising.csv")
data= df.to_numpy()
```

```python
sales_data= data[:,3]
sales_max= np.max(sales_data)
sales_max_ind= np.argmax(sales_data)
print(sales_max)
print(sales_max_ind)
```

```python
TV_data= data[:,0]
TV_mean= np.mean(TV_data)
print(TV_mean)
```

```python
sales_data= data[:,3]
sales_count= np.sum(sales_data >= 20)
print(sales_count)
```

```python
sales_cond= data[:,3] >= 15
radio_data= data[:,1]
radio_cond= radio_data* sales_cond
radio_mean= np.sum(radio_cond)/ np.sum(sales_cond)
print(radio_mean)
```

```python
news_cond= data[:,2] > np.mean(data[:,2])
sales_data= data[:,3]
sales_cond= sales_data* news_cond
print(np.sum(sales_cond))
```

```python
sales_data= data[:,3]
sales_mean= np.mean(sales_data)
score= np.where(
    sales_data < sales_mean, "Bad", np.where(sales_data > sales_mean,"Good","Average")
)
print(score[7:10])
```

```python
sales_data= data[:,3]
sales_mean= np.mean(sales_data)
sub_mean= abs(sales_data- sales_mean)
ind= np.argmin(sub_mean)
a= sales_data[ind]
score= np.where(
    sales_data < a ,"Bad", np.where(sales_data > a,"Good", "Average")
)
print(score[7:10])
```
